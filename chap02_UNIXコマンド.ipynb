## タブをスペースに置換

https://nlp100.github.io/ja/ch02.html#11-%E3%82%BF%E3%83%96%E3%82%92%E3%82%B9%E3%83%9A%E3%83%BC%E3%82%B9%E3%81%AB%E7%BD%AE%E6%8F%9B

<div style="border: 1px dashed #87dfff; background-color: #d6f4ff; padding: 4px">
タブ1文字につきスペース1文字に置換せよ．確認にはsedコマンド，trコマンド，もしくはexpandコマンドを用いよ．
</div>

Pythonで

```python
def replace_tab_to_space(filename, inplace=False):
    replaced = ""
    with open(filename, 'r') as fp:
        replaced = fp.read().replace(chr(9), ' ')
    if not inplace:
        return replaced
    else:
        with open(filename, 'w') as fp:
            fp.write(replaced)
        return None
```

```python
%%sh

head -n 1 ./data/popular-names.txt
```

```python
s = replace_tab_to_space('./data/popular-names.txt')
s[:20]
```

Pythonで文字を`'¥t'`と指定したり`chr(9)`と文字コードから変換してもタブ文字は検出してくれない。

UNIXコマンドで

**`sed`コマンド**

```python
%%sh

sed 's/	/ /g' ./data/popular-names.txt 
```

MacOSを使っていると`'¥t'`をタブに変換してくれないらしく、Ctrl + VからのTabで対処した。  
Linuxなら`'¥t'`でいけた記憶がある。

**`tr`コマンド**

```python
%%sh
cat ./data/popular-names.txt | tr '	' ' '
```

```python
%%sh
tr '	' ' ' < ./data/popular-names.txt
```

`tr`は引数にファイル名は受け取ることができない。標準入力として渡す必要があるので注意。

**`expand`コマンド**

```python
%%sh
expand -t 1 ./data/popular-names.txt
```

`expand`コマンドはタブ文字を`-t`オプションで指定した数のスペースに置き換える。

```python
replace_tab_to_space('./data/popular-names.txt', inplace=True)
```

## 1列目をcol1.txtに，2列目をcol2.txtに保存

https://nlp100.github.io/ja/ch02.html#12-1%E5%88%97%E7%9B%AE%E3%82%92col1txt%E3%81%AB2%E5%88%97%E7%9B%AE%E3%82%92col2txt%E3%81%AB%E4%BF%9D%E5%AD%98

<div style="border: 1px dashed #87dfff; background-color: #d6f4ff; padding: 4px">
各行の1列目だけを抜き出したものをcol1.txtに，2列目だけを抜き出したものをcol2.txtとしてファイルに保存せよ．確認にはcutコマンドを用いよ．
</div>

Pythonで

```python
def get_col(filename, col=0):
    col_values = []
    with open(filename, 'r') as fp:
        for line in fp.readlines():
            col_values.append(line.split(' ')[col])
    return col_values
```

```python
with open('./data/col1.txt', 'w') as fp:
    fp.write(chr(10).join(get_col('./data/popular-names.txt', col=0)))
    
with open('./data/col2.txt', 'w') as fp:
    fp.write(chr(10).join(get_col('./data/popular-names.txt', col=1)))
```

UNIXコマンドで

```python
%%sh
cut -f 1 -d ' ' ./data/popular-names.txt
```

```python
%%sh
cut -f 2 -d ' ' ./data/popular-names.txt
```

`cut`コマンドは`-d`オプションでデリミタ指定、`-f`でデリミタで区切られたセクションの何番目を表示するか指定する。

##  col1.txtとcol2.txtをマージ

https://nlp100.github.io/ja/ch02.html#13-col1txt%E3%81%A8col2txt%E3%82%92%E3%83%9E%E3%83%BC%E3%82%B8

<div style="border: 1px dashed #87dfff; background-color: #d6f4ff; padding: 4px">
12で作ったcol1.txtとcol2.txtを結合し，元のファイルの1列目と2列目をタブ区切りで並べたテキストファイルを作成せよ．確認にはpasteコマンドを用いよ．
</div>

Pythonで

```python
def concat_cols(col1_filename, col2_filename, out_filename='output.txt'):
    with open(col1_filename, 'r') as fp1, open(col2_filename, 'r') as fp2:
        s = [f'{c1.strip()}	{c2.strip()}' for c1, c2 in zip(fp1.readlines(), fp2.readlines())]
    with open(out_filename, 'w') as fp:
        fp.write(chr(10).join(s))
```

```python
concat_cols('./data/col1.txt', './data/col2.txt', './data/col1_2.txt')
```

UNIXコマンドで

```python
%%sh
paste -d '	' ./data/col1.txt ./data/col2.txt
```

`paste`という都合の良いコマンドがあった。

## ファイルをN分割する

https://nlp100.github.io/ja/ch02.html#16-%E3%83%95%E3%82%A1%E3%82%A4%E3%83%AB%E3%82%92n%E5%88%86%E5%89%B2%E3%81%99%E3%82%8B

<div style="border: 1px dashed #87dfff; background-color: #d6f4ff; padding: 4px">
自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．
</div>

とりあえずsplitの仕様に合わせる。

```python
def split_file(filename, N=2, prefix='x'):
    i_split = 1
    eof_flag = False
    with open(filename, 'r') as fp:
        while True:
            with open(f'{prefix}{i_split}', 'w') as ofp:
                for i in range(N):
                    line = fp.readline()
                    if not line:
                        eof_flag = True
                        break
                    ofp.write(line)
            i_split += 1
            if eof_flag:
                break
```

```python
split_file('./data/col1_2.txt', N=500, prefix='./data/col1_2-split-')
```

UNIXコマンドで

```python
%%sh 
split -l 500 ./data/col1_2.txt ./data/col1_2-split-x
```

## 各行を3コラム目の数値の降順にソート

https://nlp100.github.io/ja/ch02.html#18-%E5%90%84%E8%A1%8C%E3%82%923%E3%82%B3%E3%83%A9%E3%83%A0%E7%9B%AE%E3%81%AE%E6%95%B0%E5%80%A4%E3%81%AE%E9%99%8D%E9%A0%86%E3%81%AB%E3%82%BD%E3%83%BC%E3%83%88

<div style="border: 1px dashed #87dfff; background-color: #d6f4ff; padding: 4px">
各行を3コラム目の数値の逆順で整列せよ（注意: 各行の内容は変更せずに並び替えよ）．確認にはsortコマンドを用いよ（この問題はコマンドで実行した時の結果と合わなくてもよい）．

Pythonで

```python
with open('./data/popular-names.txt', 'r') as fp:
    popular_names = [line.strip().split(' ') for line in fp.readlines()]
    
popular_names.sort(key=lambda x: int(x[2]))
popular_names[:10]
```

```python
%%sh

sort -k 3 -t ' ' -n ./data/popular-names.txt | head
```

`sort`コマンド

* `-k <key>`: 何列目をキーにして並べ替えるか。 1オリジン。
* `-t`: デリミタを指定する。

## 各行の1コラム目の文字列の出現頻度を求め，出現頻度の高い順に並べる

https://nlp100.github.io/ja/ch02.html#19-%E5%90%84%E8%A1%8C%E3%81%AE1%E3%82%B3%E3%83%A9%E3%83%A0%E7%9B%AE%E3%81%AE%E6%96%87%E5%AD%97%E5%88%97%E3%81%AE%E5%87%BA%E7%8F%BE%E9%A0%BB%E5%BA%A6%E3%82%92%E6%B1%82%E3%82%81%E5%87%BA%E7%8F%BE%E9%A0%BB%E5%BA%A6%E3%81%AE%E9%AB%98%E3%81%84%E9%A0%86%E3%81%AB%E4%B8%A6%E3%81%B9%E3%82%8B

<div style="border: 1px dashed #87dfff; background-color: #d6f4ff; padding: 4px">
各行の1列目の文字列の出現頻度を求め，その高い順に並べて表示せよ．確認にはcut, uniq, sortコマンドを用いよ．
</div>

Pythonで

```python
from collections import defaultdict
name_count = defaultdict(int)

with open('./data/popular-names.txt') as fp:
    for line in fp.readlines():
        name = line.strip().split()[0]
        name_count[name.strip()] += 1
        
name_list = list(name_count.keys())
name_list.sort(key=lambda x: name_count[x], reverse=True)
```

```python
name_list
```

UNIXコマンドで

```python
%%sh

cut -d ' ' -f 1 ./data/popular-names.txt | sort | uniq -c | sort -k 1 -nr -t ' '
```

`man uniq`と実行してみると、`DESCRIPTION`のところに

<quote>Repeated lines in the input will not be detected if they are not adjacent, so it may be necessary to sort the files first.</quote>

(訳): 同じ値が複数行に渡って繰り返されていても、それらが隣接していない限りは検出されないので、予めファイルをソートしておく必要があるでしょう。

とあるように、`uniq`に入れる前にかならずソートしておかないといけない。
